```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import os

data_path = "/Users/ahovhannisyan/perovskite/perovskite_data_analysis/data"
composition_path = os.path.join(data_path, "composition.parquet")
crystal_system_path = os.path.join(data_path, "crystal_system.parquet")
entries_path = os.path.join(data_path, "entries.parquet")
general_info_path = os.path.join(data_path, "general_material_info.parquet")
```

```python
comp_df = pd.read_parquet(composition_path)
system_df = pd.read_parquet(crystal_system_path)
entries_df = pd.read_parquet(entries_path)
general_df = pd.read_parquet(general_info_path)
```

```python
df = pd.merge(comp_df, system_df, how="left", left_on="id", right_on="id")
df
```

```python
df = pd.merge(general_df, df, on=["id"], suffixes=("_x", None))
df.drop(columns=[col for col in df.columns if "_x" in col or "_y" in col], inplace=True)
df
```

```python
a_counts = df["A"].value_counts()
plt.figure(figsize=(30, 21))  # Width=12 inches, height=6 inches
a_counts.plot(kind="bar", rot=90, fontsize=20)
plt.title("Count of A Cations", fontsize=40)
plt.xlabel("A cation", fontsize=30)
plt.ylabel("Count", fontsize=30)
plt.tight_layout()
plt.show()
```

```python
b_counts = df["B"].value_counts()
plt.figure(figsize=(30, 21))  # Width=12 inches, height=6 inches
b_counts.plot(kind="bar", rot=90, fontsize=20)
plt.title("Count of B Cations", fontsize=40)
plt.xlabel("B cation", fontsize=30)
plt.ylabel("Count", fontsize=30)
plt.tight_layout()
plt.show()
```

```python
crystal_systems = df["crystal_system"].value_counts()
plt.figure(figsize=(30, 21))  # Width=12 inches, height=6 inches
crystal_systems.plot(kind="bar", rot=45, fontsize=20)
plt.title("Count of Systems", fontsize=40)
plt.xlabel("Systems", fontsize=30)
plt.ylabel("Count", fontsize=30)
plt.tight_layout()
plt.show()
```
