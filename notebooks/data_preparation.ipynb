### Handle perovskite materials

```python
def decompose_ions(_df: pd.DataFrame, ion_type: str = 'A', coefficient: bool = False):
    ion_type_lower = ion_type.lower()
    col = f'Perovskite_composition_{ion_type_lower}_ions'
    if coefficient:
        col = f'Perovskite_composition_{ion_type_lower}_ions_coefficients'
    _df[f'_{ion_type_lower}_ion_list'] = _df[col].str.split(';').apply(
        lambda x: [item.strip() for item in x if item.strip()] if isinstance(x, list) else []
    )
    max_components = _df[f'_{ion_type_lower}_ion_list'].apply(len).max()
    print(f"Max components for {ion_type}: {max_components}")

    for i in range(max_components):
        col_name = f'{ion_type}_{i + 1}'
        if coefficient:
            col_name = f'{ion_type}_{i + 1}_coef'
        _df[col_name] = _df[f'_{ion_type_lower}_ion_list'].apply(
            lambda x: x[i].strip() if i < len(x) else 0
        )

    _df = _df.drop(columns=[f'_{ion_type_lower}_ion_list'])
    return _df
```

```python
df_decomposed = df.copy()
df_decomposed = decompose_ions(df_decomposed, ion_type='A')
df_decomposed = decompose_ions(df_decomposed, ion_type='A', coefficient=True)
df_decomposed = decompose_ions(df_decomposed, ion_type='B')
df_decomposed = decompose_ions(df_decomposed, ion_type='B', coefficient=True)
df_decomposed = decompose_ions(df_decomposed, ion_type='C')
df_decomposed = decompose_ions(df_decomposed, ion_type='C', coefficient=True)
```

```python
cols = ["A_1", "A_2", "A_3", "A_4", "A_5", "B_1", "B_2", "B_3", "C_1", "C_2", "C_3"]
coef_cols = [f'{col}_coef' for col in cols]
df_decomposed[cols + coef_cols]
```

```python
def count_ions(_df: pd.DataFrame, site_prefix: str):
    site_cols = [col for col in _df.columns if col.startswith(site_prefix) and "coef" not in col]
    # Stack all columns into a single series and exclude -1
    all_ions = _df[site_cols].stack().reset_index(drop=True)
    all_ions = all_ions[all_ions != -1]
    return all_ions.value_counts()
```

```python
a_ion_counts = count_ions(df_decomposed, 'A_')
b_ion_counts = count_ions(df_decomposed, 'B_')
c_ion_counts = count_ions(df_decomposed, 'C_')
print("A ion counts:\n", a_ion_counts)
print("B ion counts:\n", b_ion_counts)
print("C ion counts:\n", c_ion_counts)
```

```python
# Define relevant columns by site
a_site_cols = [f"A_{i}" for i in range(1, 6)]
b_site_cols = [f"B_{i}" for i in range(1, 4)]
c_site_cols = [f"C_{i}" for i in range(1, 5)]

# Get value counts across each site group
a_counts = pd.concat([df_decomposed[col] for col in a_site_cols]).value_counts()
b_counts = pd.concat([df_decomposed[col] for col in b_site_cols]).value_counts()
c_counts = pd.concat([df_decomposed[col] for col in c_site_cols]).value_counts()

# Get allowed values per site (appear at least 350 times)
a_allowed = set(a_counts[a_counts >= 350].index)
b_allowed = set(b_counts[b_counts >= 350].index)
c_allowed = set(c_counts[c_counts >= 350].index)

# Ensure Cl is always allowed in C-site
c_allowed.add("Cl")

# Build a mask to keep only valid rows
def is_valid(row):
    a_valid = all(row[col] in a_allowed for col in a_site_cols)
    b_valid = all(row[col] in b_allowed for col in b_site_cols)
    c_valid = all(row[col] in c_allowed for col in c_site_cols)
    return a_valid and b_valid and c_valid

# Apply filter
df_decomposed = df_decomposed[df_decomposed.apply(is_valid, axis=1)]

# Show how many rows remain
df_decomposed.head()
```

```python
a_ion_counts = count_ions(df_decomposed, 'A_')
b_ion_counts = count_ions(df_decomposed, 'B_')
c_ion_counts = count_ions(df_decomposed, 'C_')
```
